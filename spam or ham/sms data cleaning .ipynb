```python
import pandas as pd
sms=pd.read_csv("D:\\leanring\\aids\\collage\\dataset\\spam.csv",encoding='latin-1')
sms.head()
```

CountVectorizer transforms text data into a matrix of word counts. Each row is a message, each column is a unique word, and the values show how often each word appears in each message. This helps turn text into numbers for machine learning.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english')
x=cv.fit_transform(sms['v2'])
```

```python
x
```

Natural Language Processing (NLP): Stemming and Lemmatization

- **Stemming** reduces words to their root form by removing suffixes (e.g., "playing" → "play").
- **Lemmatization** converts words to their base or dictionary form (e.g., "better" → "good").
- Both techniques help normalize text for analysis and improve model performance.

    **Common Text Preprocessing Steps:**
    1. Remove special characters
    2. Remove stop words
    3. Convert to lowercase
    4. Apply stemming or lemmatization

```python
y=sms['v1']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
```

```python
log_model.fit(x_train, y_train)
log_model.score(x_test, y_test) * 100 
```

```python
new_messages=cv.transform(new_messages)
log_model.predict(new_messages)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb_model = MultinomialNB()
nb_model.fit(x_train, y_train)
nb_model.score(x_test, y_test) * 100
```
